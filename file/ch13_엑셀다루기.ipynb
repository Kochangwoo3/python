```python
import glob
import re
import pandas as pd

excel_data_files1 = glob.glob("c:/dev/myPyCode/data/담당자별_판매량_*사원.xlsx")
# excel_data_files1

for f in excel_data_files1:
    df = pd.read_excel(f)
    # 특정 열의 값을 변경
    if(df.loc[1, '담당자'] == 'A'):
        df['담당자'] = 'Andy'
    elif(df.loc[1, '담당자'] == 'B'):
        df['담당자'] = 'Becky'
    elif(df.loc[1, '담당자'] == 'C'):
        df['담당자'] = 'Chris'

    # 파일명 변경
    f_new = re.sub(".xlsx", "2.xlsx", f)
    print(f_new)

    # 수정된 데이터를 새로운 파일명의 엑셀로 저장
    with pd.ExcelWriter(f_new, engine = 'xlsxwriter') as new_excel_file:
        df.to_excel(new_excel_file, index = False)
```
