```python
import pandas as pd
#엑셀 파일 읽기
df = pd.read_excel("D:/dev/myPyCode/data/담당자별_판매량_통합.xlsx")
#구두만 있는 행만 추출
product_name = '핸드백'
handbag = df[(df['제품명'] == product_name)]
# 행별로 합계를 구하고 열 추가
numeric_columns = handbag.select_dtypes(include=['int64', 'float64'])
handbag_sum = pd.DataFrame(numeric_columns.sum(axis=1), columns = ['연간판매량'])
handbag_total = handbag.join(handbag_sum)
# 분기별 합계 구하고 마지막 행에 추가
handbag_sum2 = pd.DataFrame(handbag_total.sum(), columns=['합계'])
handbag_total2 = pd.concat([handbag_total, handbag_sum2.T])
# 문자열 항목값 변경
handbag_total2.loc['합계', '제품명'] = product_name
handbag_total2.loc['합계', '담당자'] = '전체'
handbag_total2.loc['합계', '지역'] = '전체'
# 결과 확인
handbag_total2
```

## 엑셀 데이터의 시각화

```python
import matplotlib.pyplot as plt
import pandas as pd

sales = {'시간': [9, 10, 11, 12, 13, 14, 15],
         '제품1': [10, 15, 12, 11, 12, 14, 13],
         '제품2': [9, 11, 14, 12, 13, 10, 12]}
```

```python
df = pd.DataFrame(sales, index=sales['시간'], columns=['제품1', '제품2'])
df.index.name = '시간'
df
```

```python
import matplotlib
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
prod_plot = df.plot(grid=True, style=['-*', '-o'], title='시간대별 생산량')
prod_plot.set_ylabel("생산량")

image_file = 'D:/dev/myPyCode/figures/fig_for_excel1.png'
plt.savefig(image_file, dpi=400)
plt.show()
```

```python
import pandas as pd
# 1. ExcelWriter 객체 생성
excel_file = 'D:/dev/myPyCode/data/data_image_to_excel.xlsx'
excel_writer = pd.ExcelWriter(excel_file, engine='xlsxwriter')
# 2. 엑셀 시트에 쓰기 df
df.to_excel(excel_writer, index=True, sheet_name='Sheet1')

# 3. 워크시트 객체 생성
workbook = excel_writer.book
worksheet = excel_writer.sheets['Sheet1']
# 4. 워크시트 객체에 위치 지정 이미지 넣기
worksheet.insert_image('D2', image_file, {'x_scale':0.7, 'y_scale':0.7})
# 5. ExcelWriter 객체 닫고 엑셀 파일 출력
excel_writer.close()
```

```python
# 1. pandas 의 ExcelWriter 객체 생성
excel_chart = pd.ExcelWriter('D:/dev/myPyCode/data/data_chart_in_excel2.xlsx', engine='xlsxwriter')
# 2. df 데이터를 지정된 엑셀 시트에 쓰기
df.to_excel(excel_chart, index=True, sheet_name='Sheet1')
# 3. ExcelWriter 객체에서 워크북과 워크시트 객체 생성
workbook = excel_chart.book
worksheet = excel_chart.sheets['Sheet1']
# 4. 차트 객체 생성
chart = workbook.add_chart({'type': 'line'});

# 5. 차트 생성 위한 데이터 값의 범위 지정
chart.add_series({
    'values': '=Sheet1!$B$2:$B$8',
    'categories': '=Sheet1!$A$2:$A$8',
    'name': '=Sheet1!$B$1',
})
chart.add_series({
    'values': '=Sheet1!$C$2:$C$8',
    'categories': '=Sheet1!$A$2:$A$8',
    'name': '=Sheet1!$C$1',
})
# 5 엑셀 차트에 X, Y축 라벨과 제목 추가
chart.set_title({'name': '시간대별 생산량'})
chart.set_x_axis({'name': '시간'}) 
chart.set_y_axis({'name': '생산량'})

# 6 워크시트에 차트 넣을 위치 지정  후 close()
worksheet.insert_chart('D2', chart)
excel_chart.close();
```
